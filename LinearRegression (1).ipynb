```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
s=pd.read_csv("TM.csv")
s
```

```python
x=s.iloc[:,:-1].values
y=s.iloc[:,1].values
```

```python
x
```

```python
y
```

```python
from sklearn.linear_model import LinearRegression
regressor=LinearRegression()
regressor.fit(x,y)
y_pred=regressor.predict(x)
plt.scatter(x,y,color="red")
plt.plot(x,y_pred,color="blue")
plt.xlabel("Days")
plt.ylabel("Price")
plt.title("Price of Toyota share")
```
